```python
from plotnine import *
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['svg.fonttype'] = 'none'
```

```python
df = pd.read_csv("countries.csv")
df.head()
```

```python
# must be svg or pdf
chart = (
    ggplot(df)
    + aes(x='GDP_per_capita', y='life_expectancy', color='Continent')
    + geom_point(size=4, alpha=0.5)
    + ylim(30, 90)
    + labs(title="Here are some countries")
    + theme_bw()
    + theme(figure_size=(10, 7))
)
chart.save("countries.svg")
chart
```
